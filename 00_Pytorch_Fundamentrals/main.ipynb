## Reshaping, Viewing and Stacking Tensors

```python
x = torch.arange(1., 10.)
x, x.shape
```

```python
## Add an extra dimension 

x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1,9)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor shares the same memory)

z[:, 0] = 5 # esta forma se le asigna a la primera columna cuando se maneja más de una dimensión
x, z
```

```python
a = torch.tensor([1,2])
b = torch.tensor([3,4])
c = torch.tensor([5,6])

stacked = torch.stack([a,b,c], dim=1)
stacked 
```

## PyTorch tensors & Numpy

Pytorch has functionality to interact with it 

- Numpy -> PyTorch tensor = torch.from_numpy(ndarray)
- Pytorch tensor -> NumPy = torch.Tensor.numpy()

```python
import torch 
import numpy as np 

array = np.arange(1, 8)

tensor = torch.from_numpy(array)

array, tensor
```

```python
array.dtype, tensor.dtype
```

```python
# Change 
tensor = torch.ones(5)
numpy_tensor = tensor.numpy()

tensor, numpy_tensor

tensor = tensor + 1
tensor, numpy_tensor

```

## Pytorch Reproducibility 

start with random numbers -> tensor operations -> update random numbers to try and make them representation of the data -> again -> again -> again


```python
torch.rand(3,3)
```

```python
random_tensor_a = torch.rand(3,3)
random_tensor_b = torch.rand(3,3)

print(random_tensor_a)
print(random_tensor_b)

print(random_tensor_a == random_tensor_b)
```

```python
random_seed = 42
torch.manual_seed(random_seed)

random_tensor_r = torch.rand(3,3)

torch.manual_seed(random_seed)

random_tensor_d = torch.rand(3,3)

print(random_tensor_r == random_tensor_d)
```
